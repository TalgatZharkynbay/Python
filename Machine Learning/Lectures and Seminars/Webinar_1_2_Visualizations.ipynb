### Plot plot and basic parameters
Let's generate sine values on the segment from 0 to 3 Pi.

Note that the `plt.plot` function takes at least 2 arguments: the x-axis dataset and the y-axis dataset. The result is a set of pairs of the form:

(x1,y1), (x2,y2), ... , (xn,yn), which is displayed by the `plt.plot` function. `plt.plot` connects all our points with a line, the more points we have, the smoother the line gets

Thus the x and y dimensions must coincide.

```python
import numpy as np
```

```python
x = np.arange(0, 3 * np.pi, 0.1)
y = np.sin(x)
print("x is", x)
print("y is {}".format(y))

plt.plot(x, y)
```

The style when we operate with a global object plt ("drawer"), only changing in which cell this object is currently drawing, is very similar to the Matlab style. This style is also called functional style.

There is an alternative, `object', where we explicitly operate on the entire `figure' and `axes'.

In our case, `plt.figure` points to an object that takes such properties as figure size (figsize) or background color (facecolor).

[Full list of parameters.](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.figure.html)

In addition, other methods are available in the title object, for example, we can title the graph with `plt.title`, and we can sign the axes with `plt.xlabel` and `plt.ylabel`. All this, is a good tone when plotting graphs.

```python
# title and axes signatures
x = np.arange(0, 3 * np.pi, 0.1)
y = np.sin(x) # element-by-element calculation of sine function values

###
plt.figure(figsize=(16,9)) # width and height in inches
###
plt.plot(x, y)
###
plt.title("Sine graph")
plt.xlabel("x")
plt.ylabel("y")
###
plt.show()
```

By calling the plt.plot function several times we will add plots to the canvas.

With plt.legend we can specify a legend for our plots.

```python
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.figure(figsize=(16,9))
plt.plot(x, y_sin)
plt.plot(x, y_cos)
plt.title("Graph of sine and cosine")
plt.xlabel("x")
plt.ylabel("y")
###
plt.legend(['sin', 'cos'])
###
plt.show()
```

We can also display text in any place of our chart, for this purpose we pass the x and y coordinate of the upper left corner of the block with our text to plt.text, and then specify the text itself and properties for displaying the text.

```python
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.figure(figsize=(16,9))
plt.plot(x, y_sin)
plt.plot(x, y_cos)
plt.title("Graph of sine and cosine")
plt.xlabel("x")
plt.ylabel("y")
plt.legend(['sin', 'cos'])
###
plt.text(2, -0.5, 'Its some text on a graphic', bbox=dict(facecolor='red', alpha=0.5), fontsize=14)
###
plt.show()
```

Curious linestyle option for `plt.plot`

```python
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.figure(figsize=(16,9))
###
plt.plot(x, y_sin, linestyle='--')
plt.plot(x, y_cos, linestyle='dashdot')
###
plt.title("Graph of sine and cosine")
plt.xlabel("x")
plt.ylabel("y")
plt.legend(['sin', 'cos'])
plt.text(2, -0.5, 'Its some text on a graphic', bbox=dict(facecolor='red', alpha=0.5), fontsize=14)
plt.show()
```

You can also change the line thickness with `linewidth` and its transparency with `alpha`. Transparency is also available for other drawing functions. This option is useful when you have a significant overlap between one chart and another.

```python
# Changing line thickness and transparency
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.figure(figsize=(16,9))
###
plt.plot(x, y_sin, linestyle='--', linewidth=10, alpha=0.2)
###
plt.plot(x, y_cos, linestyle='dashdot')
plt.title("Graph of sine and cosine")
plt.xlabel("x")
plt.ylabel("y")
plt.legend(['sin', 'cos'])
plt.text(2, -0.5, 'Its some text on a graphic', bbox=dict(facecolor='red', alpha=0.5, boxstyle = 'round,pad=0.5'), fontsize=14)
plt.show()
```

You can find the full set of options for `plt.plot()` [here](https://matplotlib.org/3.3.1/api/_as_gen/matplotlib.pyplot.plot.html).

### Scatter plot

Let's consider other functions for drawing. For example, `plt.scatter` also takes a list of x and y as input, but does not connect the points to each other.

```python
plt.scatter(x, y_sin)
```

We can, for example, change the shape of a point

```python
plt.scatter(x, y_sin, marker='*',)
```

If we pass a list of the same size as x and y to option s, we will have the size of the points change according to the size we specify

```python
plt.figure(figsize=(16,9))
plt.scatter(x, y_sin, s=np.abs(y_cos)*20)
```

Similarly, we can change the color of the points using the parameter c

```python
plt.figure(figsize=(16,9))
colors_glossary = ['#39AAA67F', 'purple']


# for i in range(x.shape[0]):
#     colors.append(colors_glossary[i % 2])

colors = [colors_glossary[i % 2] for i in range(x.shape[0])]
plt.scatter(x, y_sin, s=np.abs(y_cos)*20, c=colors)
plt.grid(axis = 'x', color='black', linestyle='-.', linewidth=0.3)
```

### Functional and object approach in Matplotlib

Using the global object plt, we can draw multiple pairs of axes on a single "canvas". As already mentioned, this represents the so-called first or functional approach.

plt.subplot takes the first parameter the number of rows, the second the number of columns, and the last parameter sets the axis pair we are working with now.

I recommend practicing with this example yourself.

```python
plt.figure(figsize=(12,3))
plt.subplot(1,2,1) # from now on we draw on the pair of axes 1
#
#
plt.plot(x, y_sin)
#
#
plt.subplot(1,2,2) # from now on we draw on the pair of axes 2
#
#
plt.plot(x, y_cos)
#
#
plt.show()
```

Below we will show an alternative way to display several graphs on one canvas. Here we use the so-called "object" interface, which we have already mentioned above.

[plt.sublots()](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplots.html) also accepts the number of rows and columns, as well as all the parameters that `plt.figure()` accepts. Returns two objects: a figure object and a set of axis objects as a numpy array of dimension number of rows by number of columns.

```python
# An alternative way to display multiple charts on a single canvas
fig, axes = plt.subplots(2,2, figsize=(16,9))
print(axes.shape)
l = [[1,2,3], [4,5,6], [8,8,9], [100,1,2,3]]
for idx, ax in enumerate(axes.ravel()):
  ax.hist(l[idx])
```

### Improving and optimizing graphs

So, here are 10 moderately abstract steps that will greatly improve your visualizations and upgrade your Matplotlib skill set

1. Be clear - what is even depicted?
2. Compare - put related information side by side
3. Simplify - do we really need all the dots to convey the message?
4. Focus attention - is the whole range of values needed, does it make sense?
5. Encourage questions - you can highlight interesting answers to emerging questions right on the chart
6. Try to answer the "why" question - the graph should ideally make the idea clear without any commentary
7. Keep an opportunity to validate the visualization - you can leave a few value points with captions
8. Do different visualizations - there is no one-size-fits-all type of visualization, look for what fits best
9. Healthy skepticism - cite references, links. Ask other people if something in the graph is clear or understandable to them
10. Be detailed - the most interesting points may be worth considering separately

```python
# 7
def get_point_for_dt(date_arr: np.ndarray, value_arr: np.ndarray, dt: str): # +
  return value_arr[np.argwhere(date_arr == np.datetime64(dt)).flatten()] # +


fig, ax1 = plt.subplots(figsize=(16,9))

color = 'tab:blue'
ax1.set_xlabel('Time')
max_date = np.max(np.concatenate([btc_usd_date, money_supply_date]))
ax1.set_xlim(left=np.datetime64('2014-01-01'), right=max_date)
ax1.set_ylabel('BTC/USD price, $', color = color)
ax1.plot(btc_usd_date, btc_usd_close, color = color)
ax1.tick_params(axis ='y', labelcolor = color)

# The right y-axis would be for currency interventions
ax2 = ax1.twinx()

color = 'tab:purple'
ax2.set_ylabel('Volume of currency interventions, mln $', color = color)
ax2.plot(money_supply_date, money_supply_value, color = color)
ax2.scatter([np.datetime64('2020-02-29'), ], get_point_for_dt(money_supply_date, money_supply_value, '2020-02-29'), color=color) # +
ax2.scatter([np.datetime64('2020-05-31'), ], get_point_for_dt(money_supply_date, money_supply_value, '2020-05-31'), color=color) # +
ax2.tick_params(axis ='y', labelcolor = color)

plt.title("Currency interventions and the value of bitcoin USD")
plt.text(np.datetime64('2020-04-09'), 3.2e6, '$2.3 trillion pump', bbox=dict(facecolor='purple', alpha=0.5), fontsize=12)
plt.text(np.datetime64('2019-09-25'), # +
         3.5e6, f"{get_point_for_dt(money_supply_date, money_supply_value, '2020-02-29')[0]}",  # +
         fontsize=12) # +

plt.text(np.datetime64('2019-12-15'), # +
         5.1e6, f"{get_point_for_dt(money_supply_date, money_supply_value, '2020-05-31')[0]}", # +
         fontsize=12) # +

plt.text(money_supply_date[list(money_supply_value).index(max(money_supply_value))], # +
         max(money_supply_value), f"{get_point_for_dt(money_supply_date, money_supply_value, money_supply_date[list(money_supply_value).index(max(money_supply_value))])[0]}",  # +
         fontsize=12) # +

plt.show()
```

```python
# get correlation matrix
corr = iris.corr(numeric_only = True)
fig, ax = plt.subplots()
# create heatmap
im = ax.imshow(corr.values)

# set labels
ax.set_xticks(np.arange(len(corr.columns)))
ax.set_yticks(np.arange(len(corr.columns)))
ax.set_xticklabels(corr.columns)
ax.set_yticklabels(corr.columns)

# Rotate the tick labels and set their alignment.
plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")

# Loop over data dimensions and create text annotations.
for i in range(len(corr.columns)):
    for j in range(len(corr.columns)):
        text = ax.text(j, i, np.around(corr.iloc[i, j], decimals=2),
                       ha="center", va="center", color="black")
```
